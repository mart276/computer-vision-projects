```python
import cv2 as cv
import numpy as np


lower_white = np.array([0, 0, 100], dtype=np.uint8)
upper_white = np.array([180, 40, 200], dtype=np.uint8)

# read the webcam
cap = cv.VideoCapture(0)

# visualize the webcam
while True:
    ret, frame = cap.read()

    hsvImage = cv.cvtColor(frame, cv.COLOR_BGR2HSV)

    mask = cv.inRange(hsvImage, lower_white, upper_white)

    if ret:
        cv.imshow('webcam', frame)
        cv.imshow('mask', mask)
        cv.waitKey(10)

    if cv.waitKey(40) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
