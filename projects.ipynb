```python
import cv2 as cv
import numpy as np


lower_color = np.array([10, 50, 100], dtype=np.uint8)
upper_color = np.array([30, 155, 200], dtype=np.uint8)

# read the webcam
cap = cv.VideoCapture(0)

# visualize the webcam
while True:
    ret, frame = cap.read()

    hsvImage = cv.cvtColor(frame, cv.COLOR_BGR2HSV)

    mask = cv.inRange(hsvImage, lower_color, upper_color)

    contours, heirarchy = cv.findContours(mask, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

    _, inv = cv.threshold(mask, 127, 255, cv.THRESH_BINARY_INV)

    for ent in contours:
        if cv.contourArea(ent) > 2500:
            x1, y1, w, h = cv.boundingRect(ent)
            cv.rectangle(frame, (x1 - 10, y1 - 10), (x1 + w + 10, y1 + h + 10), (0, 255, 0), 2)

    if ret:
        cv.imshow('webcam', frame)
        # cv.imshow('mask', inv)
        cv.waitKey(10)

    if cv.waitKey(40) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
