```python
# play with images using OpenCV
# update git

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photos', 'shower.png')

img = cv.imread(image_path)

# write image
cv.imwrite(os.path.join('.', 'photos', 'shower_copy.png'), img)

# visualize the image
cv.imshow('shower', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# play with video using OpenCV

import os
import cv2 as cv

# read the video
video_path = os.path.join('.', 'videos', 'teio.mp4')

cap = cv.VideoCapture(video_path)

# visualize the video
while True:
    ret, frame = cap.read()

    if ret:
        cv.imshow('teio', frame)
        cv.waitKey(25)

    if cv.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# pley with the webcam using OpenCV

import cv2 as cv

# read the webcam
cap = cv.VideoCapture(0)

# visualize the webcam
while True:
    ret, frame = cap.read()

    if ret:
        cv.imshow('webcam', frame)
        cv.waitKey(10)

    if cv.waitKey(40) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# basic resizing with OpenCV

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photos', 'shower.png')
img = cv.imread(image_path)

resized_img = cv.resize(img, (int(img.shape[1] * 0.5), int(img.shape[0] * 0.5)))

print('Image shape:', img.shape)
print('Resized image shape:', resized_img.shape)
cv.imshow('shower', img)
cv.imshow('resized_shower', resized_img)
cv.waitKey(0)
```

```python
#  color space conversion with OpenCV

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photos', 'shower.png')
img = cv.imread(image_path)

# resize the image
ratio = float(input('Enter the resizing ratio (e.g., 0.75 for 75%): '))
resized_img = cv.resize(img, (int(img.shape[1] * ratio), int(img.shape[0] * ratio)))

# convert the image to grayscale
gray_img = cv.cvtColor(resized_img, cv.COLOR_BGR2GRAY)
bgr_img = cv.cvtColor(resized_img, cv.COLOR_RGB2BGR)
hsv_img = cv.cvtColor(resized_img, cv.COLOR_BGR2HSV)

# visualize the image
cv.imshow('original_shower', resized_img)
cv.imshow('gray_shower', gray_img)
cv.imshow('bgr_shower', bgr_img)
cv.imshow('hsv_shower', hsv_img)
cv.waitKey(0)
```

```python
# blur the image with OpenCV

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photos', 'shower.png')
img = cv.imread(image_path)

# resize the image
# resized_img = cv.resize(img, (int(img.shape[1] * 0.75), int(img.shape[0] * 0.75)))
cropped_img = img[100:400, 100:400]

# blur the image
K_SIZE = 100
cv.blur(cropped_img, (K_SIZE, K_SIZE))

# visualize the image
cv.imshow('shower', cropped_img)
cv.imshow('blurred_shower', cropped_img)
cv.waitKey(0)
```
