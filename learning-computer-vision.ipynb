```python
# play with video using OpenCV

import os
import cv2 as cv

# read the video
video_path = os.path.join('.', 'videoes', 'teio.mp4')

cap = cv.VideoCapture(video_path)

# visualize the video
while True:
    ret, frame = cap.read()

    if ret:
        cv.imshow('teio', frame)
        cv.waitKey(25)

    if cv.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# pley with the webcam using OpenCV

import cv2 as cv

# read the webcam
cap = cv.VideoCapture(0)

# visualize the webcam
while True:
    ret, frame = cap.read()

    if ret:
        cv.imshow('webcam', frame)
        cv.waitKey(10)

    if cv.waitKey(40) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
#  color space conversion with OpenCV

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photoes', 'shower.png')
img = cv.imread(image_path)

# resize the image
ratio = float(input('Enter the resizing ratio (e.g., 0.75 for 75%): '))
resized_img = cv.resize(img, (int(img.shape[1] * ratio), int(img.shape[0] * ratio)))

# convert the image to grayscale
gray_img = cv.cvtColor(resized_img, cv.COLOR_BGR2GRAY)
bgr_img = cv.cvtColor(resized_img, cv.COLOR_RGB2BGR)
hsv_img = cv.cvtColor(resized_img, cv.COLOR_BGR2HSV)

# visualize the image
cv.imshow('original_shower', resized_img)
cv.imshow('gray_shower', gray_img)
cv.imshow('bgr_shower', bgr_img)
cv.imshow('hsv_shower', hsv_img)
cv.waitKey(0)
```

```python
# blur the image with OpenCV

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photoes', 'shower.png')
img = cv.imread(image_path)

# resize the image
cropped_img = img[100:400, 100:400]
# resized_img = cv.resize(cropped_img, int(cropped_img.shape[1]*2), int(cropped_img.shape[0]*2))

# blur the image
K_SIZE = 15
blured_img = cv.blur(cropped_img, (K_SIZE, K_SIZE))
img_gaussian_blur = cv.GaussianBlur(cropped_img, (K_SIZE, K_SIZE), 3)

# visualize the image
cv.imshow('shower', cropped_img)
cv.imshow('blurred_shower', blured_img)
cv.imshow('gaussian_blured_shower', img_gaussian_blur)
cv.waitKey(0)
```

```python
# remove noises using open cv

import cv2 as cv
import os

# read image
img_path = os.path.join('.', 'photoes', 'shower.png')
img = cv.imread(img_path)

# process the image
K_size = 5

cropped_img = img[100:400, 100:400]
gray_img = cv.cvtColor(cropped_img, cv.COLOR_BGR2GRAY)

blured_img = cv.blur(gray_img, (K_size, K_size))
gaussian_blured_img = cv.GaussianBlur(gray_img, (K_size, K_size), 5)
median_blured_img = cv.medianBlur(gray_img, K_size)

# visualize the image
cv.imshow('shower', cropped_img)
cv.imshow('blured shower', blured_img)
cv.imshow('gaussian blured shower', gaussian_blured_img)
cv.imshow('median blured shower', median_blured_img)
cv.imshow('gray shower', gray_img)
cv.waitKey(0)
```

```python
# image threshold

import cv2 as cv
import os

# read image
img_path = os.path.join('.', 'photoes', 'shower.png')
img = cv.imread(img_path)

#image rpocessing
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

adaptive_thresh = cv.adaptiveThreshold(gray_img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 21, 30)
ret, simple_thresh = cv.threshold(gray_img, 80, 255, cv.THRESH_BINARY)

# visualize
cv.imshow('shower', img)
cv.imshow('simple thresholding', simple_thresh)
cv.imshow('adaptive thresholding', adaptive_thresh)
cv.waitKey(0)
```
