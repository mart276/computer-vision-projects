```python
# blur the image with OpenCV

import os
import cv2 as cv

# read the image
image_path = os.path.join('.', 'photoes', 'shower.png')
img = cv.imread(image_path)

# resize the image
cropped_img = img[100:400, 100:400]
# resized_img = cv.resize(cropped_img, int(cropped_img.shape[1]*2), int(cropped_img.shape[0]*2))

# blur the image
K_SIZE = 15
blured_img = cv.blur(cropped_img, (K_SIZE, K_SIZE))
img_gaussian_blur = cv.GaussianBlur(cropped_img, (K_SIZE, K_SIZE), 3)

# visualize the image
cv.imshow('shower', cropped_img)
cv.imshow('blurred_shower', blured_img)
cv.imshow('gaussian_blured_shower', img_gaussian_blur)
cv.waitKey(0)
```

```python
# remove noises using open cv

import cv2 as cv
import os

# read image
img_path = os.path.join('.', 'photoes', 'shower.png')
img = cv.imread(img_path)

# process the image
K_size = 5

cropped_img = img[100:400, 100:400]
gray_img = cv.cvtColor(cropped_img, cv.COLOR_BGR2GRAY)

blured_img = cv.blur(gray_img, (K_size, K_size))
gaussian_blured_img = cv.GaussianBlur(gray_img, (K_size, K_size), 5)
median_blured_img = cv.medianBlur(gray_img, K_size)

# visualize the image
cv.imshow('shower', cropped_img)
cv.imshow('blured shower', blured_img)
cv.imshow('gaussian blured shower', gaussian_blured_img)
cv.imshow('median blured shower', median_blured_img)
cv.imshow('gray shower', gray_img)
cv.waitKey(0)
```

```python
import cv2 as cv
import numpy as np

# read the webcam
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    b, g, r = cv.split(frame)

    def frame_mod(channels):
        mod_channels = []
        for channel in channels:
            _, thresh = cv.threshold(channel, 15, 255, cv.THRESH_TOZERO)
            blurred = cv.medianBlur(thresh, 5)
            mod_channels.append(blurred)
        return cv.merge(mod_channels)
    
    def frame_edge(frame):
        edge = cv.Canny(frame, 40, 60)
        dilated_edge = cv.dilate(edge, np.ones((2, 2), dtype=np.uint8))
        blured = cv.medianBlur(dilated_edge, ksize=5)
        return blured, dilated_edge

    modded_frame = frame_mod([b, g, r])
    edge_frame, dilated_edge = frame_edge(frame)

    # cv.imshow('original', frame)
    # cv.imshow('modded', modded_frame)
    cv.imshow('dilated_edge', dilated_edge)
    cv.imshow('edge', edge_frame)

    if cv.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import os

img = cv.imread(os.path.join('.', 'photoes', 'birds.jpg'))
crop = img[0:290, 0:612]
gray = cv.cvtColor(crop, cv.COLOR_BGR2GRAY)
ret, thres = cv.threshold(gray, 20, 255, cv.THRESH_BINARY_INV)

contours, hierarchy = cv.findContours(thres, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

for ent in contours:
    if cv.contourArea(ent) > 5:
        cv.drawContours(img, ent, -1, (0, 255, 0), 1)

# cv.imshow('image', crop)
cv.imshow('thres', thres)
# print(thres.shape)
cv.waitKey(0)
```
